```python
data.replace({"Item_Fat_Content":{"low fat" : "Low Fat","LF" : "Low Fat","reg" : "Regular"}},inplace=True)
```

```python
data['Item_Fat_Content'].unique()
```

```python
data
```

```python

colms = ['Item_Fat_Content', 'Item_Type', 'Outlet_Size', 'Outlet_Location_Type', 'Outlet_Type','Item_Identifier','Outlet_Identifier']

encoder = LabelEncoder()
for col in colms:
    data[col] = encoder.fit_transform(data[col])
```

```python
data.head()
```

```python
X = data.drop(columns='Item_Outlet_Sales',axis=1)
y = data['Item_Outlet_Sales']
```

```python
X
```

```python
y
```

```python
X_train , X_test , y_train,y_test = train_test_split(X,y,test_size=0.25,shuffle=True,random_state=7)
```

```python
models = [XGBRFRegressor(),LinearRegression(),KNeighborsRegressor()]

for model in models:
    regressor = model
    print("Model : ",regressor)
    regressor.fit(X_train,y_train)
    X_train_pred = regressor.predict(X_train)
    X_test_pred = regressor.predict(X_test)
    r2_score_train = metrics.r2_score(y_train,X_train_pred)
    print("R2 Square score on Train Data : ", r2_score_train*100)
    r2_score_test = metrics.r2_score(y_test,X_test_pred)
    print("R2 Square score on Test Data : ", r2_score_test*100)
```
